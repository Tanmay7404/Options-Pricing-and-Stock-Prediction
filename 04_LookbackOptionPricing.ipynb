```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt

T=1
Sigma=np.linspace(0.1,0.3,100)
R=np.linspace(0.04,0.12,100)
K=np.linspace(50,150,100,dtype=int)
S_0=np.linspace(50,150,100,dtype=int)
M=np.linspace(50,150,100,dtype=int)
def up(t,sigma,r):
    return math.exp(sigma *math.sqrt(t) +(r -0.5 * sigma**2) *t)

def down(t, sigma,  r):
    return math.exp(-sigma* math.sqrt(t)+ (r -0.5 * sigma**2) *t)

def up_1(t,sigma,r):
    return math.exp(sigma *math.sqrt(t))

def down_1(t, sigma,  r):
    return math.exp(-sigma* math.sqrt(t))


def nCR(n, r):
    return math.comb(n,r)

def s(s_0,c1,c2,u,d):
    return s_0* (u**c1) *(d**c2)

def option_pricing(m,T,s_0,k,r,sigma):
    delta_t= T/ m
    u =up(delta_t, sigma, r)
    d =down(delta_t, sigma, r)
    risk_p =(math.exp(r * delta_t)- d)/(u - d)

    option_price_call= 0
    option_price_put= 0


    for i in range(0,m+ 1):
        s_price = s(s_0,m - i,i,u,d)
        payoff_call = max(0,s_price - k)
        payoff_put =max(0,k - s_price)
        prob = nCR(m, i)* (risk_p**i) *((1 - risk_p)**(m - i))
        option_price_call+= payoff_call* prob
        option_price_put+= payoff_put* prob

    dis_fac=math.exp(-r * T)
    return option_price_call * dis_fac, option_price_put * dis_fac


def option_pricing_2(m,T,s_0,k,r,sigma):
    delta_t= T/ m
    u =up_1(delta_t, sigma, r)
    d =down_1(delta_t, sigma, r)
    risk_p =(math.exp(r * delta_t)- d)/(u - d)

    option_price_call= 0
    option_price_put= 0


    for i in range(0,m+ 1):
        s_price = s(s_0,m - i,i,u,d)
        payoff_call = max(0,s_price - k)
        payoff_put =max(0,k - s_price)
        prob = nCR(m, i)* (risk_p**i) *((1 - risk_p)**(m - i))
        option_price_call+= payoff_call* prob
        option_price_put+= payoff_put* prob

    dis_fac=math.exp(-r * T)
    return option_price_call * dis_fac, option_price_put * dis_fac
arr_call=[]
arr_put=[]
for sigma in Sigma:
    x,y=option_pricing(100,1,100,100,0.08,sigma)
    arr_call.append(x)
    arr_put.append(y)

plt.figure(figsize=(10,5))
plt.title("set 1")
plt.plot(M, arr_call, label="Call Option Price")
plt.plot(M, arr_put, label="Put Option Price")
plt.legend()
plt.show()

arr_call=[]
arr_put=[]

for sigma in Sigma:
    x,y=option_pricing_2(100,1,100,100,0.08,sigma)
    arr_call.append(x)
    arr_put.append(y)

plt.figure(figsize=(10,5))
plt.title("set 2")
plt.plot(M, arr_call, label="Call Option Price")
plt.plot(M, arr_put, label="Put Option Price")
plt.legend()
plt.show()



arr_call=[]
arr_put=[]
for r in R:
    x,y=option_pricing(100,1,100,100,r,0.2)
    arr_call.append(x)
    arr_put.append(y)

plt.figure(figsize=(10,5))
plt.title("set 1")
plt.plot(M, arr_call, label="Call Option Price")
plt.plot(M, arr_put, label="Put Option Price")
plt.legend()
plt.show()


arr_call=[]
arr_put=[]
for r in R:
    x,y=option_pricing_2(100,1,100,100,r,0.2)
    arr_call.append(x)
    arr_put.append(y)

plt.figure(figsize=(10,5))
plt.title("set 2")
plt.plot(M, arr_call, label="Call Option Price")
plt.plot(M, arr_put, label="Put Option Price")
plt.legend()
plt.show()

arr_call=[]
arr_put=[]
for k in K:
    x,y=option_pricing(100,1,100,k,0.08,0.2)
    arr_call.append(x)
    arr_put.append(y)

plt.figure(figsize=(10,5))
plt.title("set 1")
plt.plot(M, arr_call, label="Call Option Price")
plt.plot(M, arr_put, label="Put Option Price")
plt.legend()
plt.show()


arr_call=[]
arr_put=[]
for k in K:
    x,y=option_pricing_2(100,1,100,k,0.08,0.2)
    arr_call.append(x)
    arr_put.append(y)

plt.figure(figsize=(10,5))
plt.title("set 2")
plt.plot(M, arr_call, label="Call Option Price")
plt.plot(M, arr_put, label="Put Option Price")
plt.legend()
plt.show()

arr_call=[]
arr_put=[]
for s_0 in S_0:
    x,y=option_pricing(100,1,s_0,100,0.08,0.2)
    arr_call.append(x)
    arr_put.append(y)

plt.figure(figsize=(10,5))
plt.title("set 1")
plt.plot(M, arr_call, label="Call Option Price")
plt.plot(M, arr_put, label="Put Option Price")
plt.legend()
plt.show()

arr_call=[]
arr_put=[]
for s_0 in S_0:
    x,y=option_pricing_2(100,1,s_0,100,0.08,0.2)
    arr_call.append(x)
    arr_put.append(y)

plt.figure(figsize=(10,5))
plt.title("set 2")
plt.plot(M, arr_call, label="Call Option Price")
plt.plot(M, arr_put, label="Put Option Price")
plt.legend()
plt.show()

arr_call=[]
arr_put=[]

for m in M:
    x,y=option_pricing(m,1,100,100,0.08,0.2)
    arr_call.append(x)
    arr_put.append(y)

plt.figure(figsize=(10,5))
plt.title("set 1")
plt.plot(M, arr_call, label="Call Option Price")
plt.plot(M, arr_put, label="Put Option Price")
plt.legend()
plt.show()


arr_call=[]
arr_put=[]

for m in M:
    x,y=option_pricing_2(m,1,100,100,0.08,0.2)
    arr_call.append(x)
    arr_put.append(y)

plt.figure(figsize=(10,5))
plt.title("set 2")
plt.plot(M, arr_call, label="Call Option Price")
plt.plot(M, arr_put, label="Put Option Price")
plt.legend()
plt.show()

```

```python
s0=100
T=1
r=0.08
sigma=0.2
M = [5,10,25,50]
```

```python
def pc(s0,T,r,sigma,m,dp,i,j,a,b,u,d,p):
    if(i==m):
        dp[i][j][a][b]=s0*(u**(a-b))*(d**b)- s0*(u**(i-j))*(d**j)
        return dp[i][j][a][b]
    pt=0
    if(dp[i][j][a][b]==-1):
        if(s0*(u**(a-b))*(d**(b))<s0*(u**(i+1-j))*(d**(j))):
            ph=pc(s0,T,r,sigma,m,dp,i+1,j,i+1,j,u,d,p)
        else:
            ph=pc(s0,T,r,sigma,m,dp,i+1,j,a,b,u,d,p)
        
        if(s0*(u**(a-b))*(d**(b))<s0*(u**(i-j))*(d**(j+1))):
            pt=pc(s0,T,r,sigma,m,dp,i+1,j+1,i+1,j+1,u,d,p)
        else:
            pt=pc(s0,T,r,sigma,m,dp,i+1,j+1,a,b,u,d,p)
        dp[i][j][a][b]=(ph*p+pt*(1-p))/(np.exp(r*(T/m)))
        return dp[i][j][a][b]
    else:
        return dp[i][j][a][b]
for m in M:
    dp = np.full((m+1,m+1,m+1,m+1),-1.0)
    dt=T/m
    u=np.exp(sigma*(np.sqrt(dt))+(r-0.5*sigma*sigma)*dt)
    d=np.exp(-sigma*(np.sqrt(dt))+(r-0.5*sigma*sigma)*dt)
    p=(np.exp(r*dt)-d)/(u-d)
    price=pc(s0,T,r,sigma,m,dp,0,0,0,0,u,d,p)
    print(f"price for M={m} is :",price)
```

As M increases (i.e., the number of subintervals in the binomial tree grows), the option price at t=0 converges to a more accurate approximation of the theoretical value derived from continuous-time models.
This is because a finer discretization (larger M) reduces the approximation error.

algorithm used in question 2 can not solve for values beyond approx 25 this is because it uses exponential algorithm O(2^M)

algorithm used in question 3 uses markov based approach to store the maximum stock price. It has time complexity of O(M^4) which can handle relatively larger values of M.
